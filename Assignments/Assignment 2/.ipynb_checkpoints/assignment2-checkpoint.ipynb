```python
import pandas as pd
import numpy as np
from IPython.display import display

cleanedDF = pd.read_csv(filepath_or_buffer = 'train.csv',
                  engine='python',
                  sep=',',
                  header=0, 
                  names=['PassengerID', 'Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked'], # the names we want to use to index columns
                  index_col=['PassengerID'], # indexing the columns
                  na_values={'PassengerID': ['NaN'],
                             'Survived': ['NaN'],
                             'Pclass': ['NaN'],
                             'Name': ['Na'],
                             'Sex': ['NA'],
                             'Age': ['NaN'],
                             'SibSp': ['NaN'],
                             'Parch': ['NaN'],
                             'Ticket': ['Na'],
                             'Fare': ['NaN'], 
                             'Cabin': ['Na'],
                             'Embarked': ['Na']} 
                 )
cleanedDF.sample(n=10, replace=True)
```

```python
# Find out which columns have missing values. 
for c in cleanedDF.columns:
    if cleanedDF[c].isnull().sum():
        print(c)
        print(cleanedDF[c].isnull().sum())

cleanedDF.info()

means = cleanedDF.groupby(["Sex","SibSp","Parch"])['Age'].transform("mean") #need to use a series to fill na
cleanedDF["Age"].fillna(means, inplace=True) #use the series from above to impute the original series. Some NaN values are still there, so further imputing by fewer group conditions is still needed.
print("Remaining missing age values:", cleanedDF["Age"].isnull().sum(), "\n")
cleanedDF['Age'] = cleanedDF['Age'].fillna(cleanedDF['Age'].mean()).round(1)
print("Remaining missing age values:", cleanedDF["Age"].isnull().sum(), "\n")
```

```python
# Drop the Cabin column because 77% of the data is missing!
cleanedDF = cleanedDF.drop('Cabin', axis=1)
cleanedDF.columns
```

```python
cleanedDF.dropna(subset=['Embarked'], inplace=True) # drop rows that do not have 'Embarked' because the number of missing ros is less than 10%
cleanedDF["Embarked"].isnull().sum() # to make sure that the rows are dropped.
```

```python
#Define the age bins and labels
age_bins = [0, 10, 20, 30, 40, 50, 60,70,130]  # Adjust the bin edges as needed
age_labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70+']
cleanedDF['AgeGroup'] = pd.cut(cleanedDF['Age'], bins=age_bins, labels=age_labels, right=False)

#Define the Fare bins and labels
fare_bins = [0, 40, 80, 120, 160, 200, 240, 280, 600]  # Adjust the bin edges as needed
fare_labels = ['0-39', '40-79', '80-119', '120-159', '160-199', '200-239', '240-279', '280+']
cleanedDF['FareGroup'] = pd.cut(cleanedDF['Fare'], bins=fare_bins, labels=fare_labels, right=False)

# Drop these columns because they are either irrelavent (Name) or have been categorized into buckets (Age).
cleanedDF = cleanedDF.drop(['Name', 'Ticket', 'Age', 'Fare'], axis=1)

# Now we have a clean datafram to analyze
cleanedDF
```

```python
# This is the "First" method described in Question 1. I 
grouped_survived = cleanedDF.groupby(["Sex", "AgeGroup", "Pclass", "SibSp", "Parch", "Embarked", "FareGroup"], observed=False)['Survived'].sum().reset_index()
grouped_total = cleanedDF.groupby(["Sex", "AgeGroup", "Pclass", "SibSp", "Parch", "Embarked", "FareGroup"], observed=False)['Survived'].count().reset_index()
grouped_survival_rate = grouped_survived['Survived'] / grouped_total['Survived']
grouped_survived
max_indices = grouped_survival_rate[grouped_survival_rate == grouped_survival_rate.max()].index
max_category_values = grouped_survived.loc[max_indices, ['Sex', 'AgeGroup', 'Pclass', "SibSp", "Parch", "Embarked", "FareGroup"]]
print(max_category_values)

grouped_survived = cleanedDF.groupby(["Sex", "AgeGroup"], observed=False)['Survived'].sum().reset_index()
grouped_total = cleanedDF.groupby(["Sex", "AgeGroup"], observed=False)['Survived'].count().reset_index()
grouped_survival_rate = grouped_survived['Survived'] / grouped_total['Survived']
grouped_survived
max_indices = grouped_survival_rate[grouped_survival_rate == grouped_survival_rate.max()].index
max_category_values = grouped_survived.loc[max_indices, ['Sex', 'AgeGroup']]
print(max_category_values)

grouped_survived = cleanedDF.groupby(["Sex", "AgeGroup", "Embarked"], observed=False)['Survived'].sum().reset_index()
grouped_total = cleanedDF.groupby(["Sex", "AgeGroup", "Embarked"], observed=False)['Survived'].count().reset_index()
grouped_survival_rate = grouped_survived['Survived'] / grouped_total['Survived']
grouped_survived
max_indices = grouped_survival_rate[grouped_survival_rate == grouped_survival_rate.max()].index
max_category_values = grouped_survived.loc[max_indices, ['Sex', 'AgeGroup', "Embarked"]]
print(max_category_values)

grouped_survived = cleanedDF.groupby(["Sex", "Pclass"], observed=False)['Survived'].sum().reset_index()
grouped_total = cleanedDF.groupby(["Sex", "Pclass"], observed=False)['Survived'].count().reset_index()
grouped_survival_rate = grouped_survived['Survived'] / grouped_total['Survived']
grouped_survived
max_indices = grouped_survival_rate[grouped_survival_rate == grouped_survival_rate.max()].index
max_category_values = grouped_survived.loc[max_indices, ['Sex', 'Pclass']]
print(max_category_values)
```

```python
# Analyze all of the columns ony by one and find the highest rate of survival in each category.
print ("Final analysis of question at hand:")
for c in cleanedDF.columns:
    if c != "Survived": # Skip the Survived column.
        grouped_survived = cleanedDF.groupby(c, observed=False)['Survived'].sum().reset_index()
        grouped_total = cleanedDF.groupby(c, observed=False)['Survived'].count().reset_index()
        grouped_survival_rate = grouped_survived['Survived'] / grouped_total['Survived'] # Find the Survival rate for this group

        # Find the indices of the highest survival rates (multiple if the same)
        max_indices = grouped_survival_rate[grouped_survival_rate == grouped_survival_rate.max()].index
        # Find the values and print.
        max_category_values = grouped_survived.loc[max_indices, [c]]
        print(max_category_values)
```
