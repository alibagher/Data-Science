```python
### Question 1:

import yfinance as yf
from pandas_datareader import data as pdr

start = datetime.datetime(2019, 3, 25)

# 1) Using pandas datareader and Yahoo Finance
yf.pdr_override()

amzn = pdr.get_data_yahoo('AMZN', start = start)
```

```python
# If we want to download multipe stocks into a single dataframe:

all_stocks_list = ['AMZN', 'META', "MMM", "IBM"]
all_stocks = yf.download(all_stocks_list, start = start)
all_stocks
```

```python
## TYPE YOUE CODE BELOW ##############################################
```

```python
### Question 2: 
amznAdjClose = all_stocks["Adj Close"].AMZN
METAAdjClose = all_stocks["Adj Close"].META
MMMAdjClose = all_stocks["Adj Close"].MMM
IBMAdjClose = all_stocks["Adj Close"].IBM
print (amznAdjClose, METAAdjClose, MMMAdjClose, IBMAdjClose)
```

```python
### Question 3:
all_stock_BME = all_stocks.resample('BM').asfreq()
amznAdjCloseBME = all_stock_BME["Adj Close"].AMZN.dropna()
METAAdjCloseBME = all_stock_BME["Adj Close"].META.dropna()
MMMAdjCloseBME = all_stock_BME["Adj Close"].MMM.dropna()
IBMAdjCloseBME = all_stock_BME["Adj Close"].IBM.dropna()
print (amznAdjCloseBME, METAAdjCloseBME, MMMAdjCloseBME, IBMAdjCloseBME)
```

```python
### Question 4:
import matplotlib.pyplot as plt  # for visualizing our data

pd.plotting.autocorrelation_plot(amznAdjCloseBME, label='AMZN')
pd.plotting.autocorrelation_plot(METAAdjCloseBME, label='META')
pd.plotting.autocorrelation_plot(MMMAdjCloseBME, label='MMM')
pd.plotting.autocorrelation_plot(IBMAdjCloseBME, label='IBM')
plt.suptitle("An Autocorrelation Graph of Adjusted Month-End Close Prices")
leg = plt.legend(loc="upper right")
plt.xlabel("Lag(Months)")
None
```

Are they autocorrelated?
- **Answer:** They are all somewhat autocorrelated.

Provide short explanation.
- **Answer:** If the plot is outside of the dashed lines, it is statistically significant and shows strong autocorrelation.
- - **AMZN**: There is a strong negative autocorrelation at around month 19-20, a medium to strong positive autocorrelation at around month 37, and a weak negative correlation at month 50.
  - **META**: There is a strong negative autocorrelation at around month 15, a medium to strong positive autocorrelation at around month 29, and a weak negative correlation at month 45-50.
  - **MMM**: There is a medium negative autocorrelation at around month 20-30, and a weak negative autocorrelation at around month 50.
  - **IBM**: There is a medium to strong negative autocorrelation at around month 35-40.

```python
### Question 5:
amznMonthlyReturn = (amznAdjCloseBME - amznAdjCloseBME.shift(1)).dropna()
METAMonthlyReturn = (METAAdjCloseBME - METAAdjCloseBME.shift(1)).dropna()
MMMMonthlyReturn = (MMMAdjCloseBME - MMMAdjCloseBME.shift(1)).dropna()
IBMMonthlyReturn = (IBMAdjCloseBME - IBMAdjCloseBME.shift(1)).dropna()
pd.plotting.autocorrelation_plot(amznMonthlyReturn, label='AMZN')
pd.plotting.autocorrelation_plot(METAMonthlyReturn, label='META')
pd.plotting.autocorrelation_plot(MMMMonthlyReturn, label='MMM')
pd.plotting.autocorrelation_plot(IBMMonthlyReturn, label='IBM')
plt.suptitle("Autocorrelation Graph of monthly returns for AMZN, META, MMM, and IBM")
plt.xlabel("Lag(Months)")
```

Are the returns autocorrelated? Provide short explanation.

**Answer:** None of the four plots are autocorrelated. The graph looks stationary meaning the lags are closer to random noise rather than showing any correlations.

```python
### Question 6:
from pandas.plotting import scatter_matrix
allReturns = pd.concat([amznMonthlyReturn, METAMonthlyReturn, MMMMonthlyReturn, IBMMonthlyReturn], axis=1)
pd.plotting.scatter_matrix(allReturns)
```
